```
A = np.array(nx.adjacency_matrix(LastFm_Graph).todense())
```

```
n=A.shape[0]
```

```

theta = 1.0/50


#idx = [7240,7572, 5993, 2381, 3767, 4640, 1516, 6973, 291, 3073]
idx = [7549, 1643, 873, 884, 885, 3764, 2212, 4622, 3706, 5478]
initActive = np.zeros((n,), dtype=bool)
initActive[idx] = 1
```

```
# Influence propagation simulation
def InfluenceProp(A, intiActive, theta, itersNum = np.inf):
    deg = np.sum(A,axis=0, dtype=float)
    i = 1 # iteration
    #i_list =[]
    resActive = initActive.copy()    
    while i < itersNum:
        i+=1
        #i_list.append(i)
        # currently inactive nodes
        inactiveId = np.where(resActive == 0)[0]    
        # LINEAR THRESHOLD MODEL
        idx = np.sum(A[np.ix_(resActive==1, resActive==0)], axis=0) / deg[resActive==0] > theta
        if np.any(idx):
            resActive[inactiveId[idx]] = 1
        else:
            break

    return resActive
```

```
# Demonstration
def ShowIteration(LastFm_Graph, initActive, resultActive, pos):
    fig = plt.figure(figsize=(15,10))
    plt.subplot(1,2,1)
    nx.draw(LastFm_Graph, pos=pos,
            nodelist=np.where(initActive)[0].tolist(),
            node_color = 'r')
    nx.draw(LastFm_Graph, pos=pos,
            nodelist=np.where(1-initActive)[0].tolist(),
            node_color = 'b')

    plt.subplot(1,2,2)
    nx.draw(LastFm_Graph, pos=pos,
            nodelist=np.where(resultActive)[0].tolist(),
            node_color = 'r')
    nx.draw(LastFm_Graph, pos=pos,
            nodelist=np.where(1-resultActive)[0].tolist(),
            node_color = 'b')
```

```
# Run 
resultActive = InfluenceProp(A, initActive, theta, itersNum = 9)

# Look 
ShowIteration(LastFm_Graph, initActive, resultActive, pos)
```
